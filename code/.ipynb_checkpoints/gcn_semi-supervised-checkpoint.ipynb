```python
import numpy as np
import keras as k
#print(k.__version__)
from keras.layers import ReLU as relu
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import random
```

Next, we need features for every node

```python
X = np.matrix([
            [i, -i]
            for i in range(4)
        ], dtype=float)
X
```

 adjacency matrix representation

```python
A = np.matrix([
    [0, 1, 0, 0],
    [0, 0, 1, 1], 
    [0, 1, 0, 0],
    [1, 0, 1, 0]],
    dtype=float
)
A
```

Adding self-loop 

(Since the node is now a neighbor of itself, the node’s own features is included when summing up the features of its neighbors!)

```python
I = np.matrix(np.eye(4))
A_hat = A + I
A_hat
```

Normalizing the Feature Representations 

(The idea of this weighted average is that we assume low-degree nodes would have bigger impacts on their neighbors, whereas high-degree nodes generate lower impacts as they scatter their influence at too many neighbors.)

```python
D = np.array(np.sum(A, axis=0))[0]
D = np.matrix(np.diag(D))
D
```

```python
D**-1 * A * X   
```

```python
W = np.matrix([
             [1, -1],
             [-1, 1]
         ]) 
D_hat**-1 * A_hat * X * W            #adding back the weihts
```

```python
relu(D_hat**-1 * A_hat * X * W)               #adding the activation function
```
